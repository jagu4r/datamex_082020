#### Show visually the frequency distribution (histogram) of the employee dataset. In few words describe these histograms?

In order to make a concise analysis we need to study salary apart from years, then we will integrate both into a conclusion.
We will first get the bins.
To get the bins we need to take into account the sample size (n = 9) and the range of the data (r = 40).
As n is a very short number and the range of salary is 40 we can separate bins into four.

```python
employee['Salary'].hist(bins = 4).set_xlabel('Salary')
employee['Salary'].hist(bins = 4).set_ylabel('Frequency')
employee['Salary'].hist(bins = 4).set_title('Salary Histogram')
```

It is set to also 4 bins in order to make a representative histogram where it is notable a salary gap.

```python
employee['Years'].hist(bins = 4).set_xlabel('Years')
employee['Years'].hist(bins = 4).set_ylabel('Frequency')
employee['Years'].hist(bins = 4).set_title('Salary Histogram')
```
